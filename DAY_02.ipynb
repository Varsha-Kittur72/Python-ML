```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.arange(1,6)
c=np.zeros((3))
d=np.ones((3,2))

print("Array a:",a)
print("Array b:",b)
print("2*3 zeros:",c)
print("3*2 ones:",d)
```

```python
print("First element of a:",a[0])
print("last two elements of a:",a[-2:])
```

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard deviation:",np.std(a))
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Department': ['Sales', 'Engineering', 'HR', 'Engineering', 'Sales'],
    'Salary': [50000, 80000, 40000, 90000, 45000]
}
df = pd.DataFrame(data)
df
```

```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

```python
print("People with salary > 50000:")
df[df['Salary'] > 50000]
```

```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

```python
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
grouped
```

```python
df['Salary in Lakhs'] = df['Salary'] / 100000
df
```

```python
import matplotlib.pyplot as plt

years = [2019, 2020, 2021, 2022]
revenue = [100, 120, 130, 160]

plt.plot(years, revenue, marker='o', linestyle='-', color='blue')
plt.title("Company Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Revenue (in Crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

```python
departments = df['Department'].value_counts()
departments.plot(kind='bar', color='teal')
plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
ages = df['Age']
plt.hist(ages, bins=5, color='coral', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('Department').size().plot(kind='pie', autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day', data=tips, palette='Set2', hue='day', legend=False)
plt.title("Count of Records by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
